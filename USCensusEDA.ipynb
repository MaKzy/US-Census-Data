#### Bi Variant Analysis

```python
col_bi=list(censusdata.columns.values)
col_bi = ['age', 'workclass', 'fnlwgt', 'education', 'education.num', 'marital.status', 'occupation', 
'relationship', 'race', 'sex', 'capital.gain', 'capital.loss', 'hours.per.week', 'native.country']
```

```python
plt.figure(figsize=(25,30))

for i in range(len(col_bi)):
    plt.subplot(5,3,i+1)
    sns.stripplot(x=censusdata['income'],y=censusdata[col_bi[i]])
    plt.title(f"Income Vs {col_bi[i]}",fontsize=20)
    plt.xticks(rotation=90,fontsize=13)
    plt.yticks(fontsize=16)
    plt.tight_layout()
```

```python
plt.figure(figsize=(8,15))
sns.stripplot(x=censusdata['income'],y=censusdata['native.country'])
#sns.stripplot(x='income',y='native.country',data=censusdata)
plt.title(f"Income Vs Native.country",fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=14)
```

```python
plt.figure(figsize=(12,8))
sns.boxenplot (x='income',y='hours.per.week',data=censusdata)
```

#### Multi - Variant Analysis

```python
col_multi = ['income', 'workclass', 'fnlwgt', 'education', 'education.num', 'marital.status', 'occupation', 
'relationship', 'race', 'sex', 'capital.gain', 'capital.loss', 'hours.per.week', 'native.country']
```

```python
plt.figure(figsize=(25,30))

for i in range(len(col_multi)):
    plt.subplot(8,2,i+1)
    sns.stripplot(x=censusdata['age'],y=censusdata[col_multi[i]])
    plt.title(f"Age Vs {col_multi[i]}",fontsize=20)
    plt.xticks(rotation=90,fontsize=13)
    plt.yticks(fontsize=16)
    plt.tight_layout()
```

```python
hoursperage= censusdata.groupby(['age','income']).apply(lambda x:x['hours.per.week'].count()).reset_index(name='hours')
hoursperage
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='age',y='hours',data=hoursperage,hue='income')
plt.title('Income Vs Work of Hours',fontsize = 30)
```
